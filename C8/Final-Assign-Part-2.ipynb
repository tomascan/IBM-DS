## TASK 2.4: Creating Callbacks; Define the callback function to update the input container based on the selected statistics and the output container

```python
## TASK 2.4: Creating Callbacks; Define the callback function to update the input container based on the selected statistics and the output container
import requests
import io
import pandas as pd
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px

# URL del archivo CSV
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"

# Obtener los datos de la URL
response = requests.get(URL)
response.raise_for_status()  # Levantar un error si la solicitud falló

# Convertir el contenido de la respuesta en un formato legible para pandas
csv_content = io.StringIO(response.text)

# Leer los datos CSV en un dataframe de pandas
df = pd.read_csv(csv_content)

# Confirmación de que los datos se han descargado y leído
print('¡Datos descargados y leídos en un dataframe!')

# Función para crear gráficos utilizando los datos
def create_graph(data, plot_id):
    fig = px.bar(data, x='Year', y='Automobile_Sales', title=f'Sales for {plot_id}')
    return fig

# Crear la aplicación Dash
app = dash.Dash(__name__)

# Lista de años para el menú desplegable de años
year_list = [i for i in range(1980, 2024)]

# Definir el layout de la aplicación
app.layout = html.Div(
    children=[
        # Título de la aplicación con estilo
        html.H1(
            "Automobile Sales Statistics Dashboard",
            style={
                'textAlign': 'center',
                'color': '#503D36',
                'font-size': '24px'
            }
        ),
        
        # Menú desplegable para seleccionar el tipo de informe
        dcc.Dropdown(
            id='dropdown-statistics',
            options=[
                {'label': 'Yearly Statistics', 'value': 'Yearly Statistics'},
                {'label': 'Recession Period Statistics', 'value': 'Recession Period Statistics'}
            ],
            placeholder='Select a report type',
            value='Select Statistics',
            style={
                'width': '80%',
                'padding': '3px',
                'font-size': '20px',
                'text-align-last': 'center'
            }
        ),
        
        # Menú desplegable para seleccionar el año
        dcc.Dropdown(
            id='select-year',
            options=[{'label': str(i), 'value': str(i)} for i in year_list],
            placeholder='Select year',
            value='Select-year',
            style={
                'width': '80%',
                'padding': '3px',
                'font-size': '20px',
                'text-align-last': 'center'
            }
        ),
        
        # División para mostrar la salida (output)
        html.Div(
            [
                html.Div(
                    id='output-container',
                    className='chart-grid',
                    style={'display': 'flex'}
                )
            ]
        )
    ]
)

# 1. Callback para actualizar el contenedor de entrada
@app.callback(
    Output(component_id='select-year', component_property='disabled'),
    Input(component_id='dropdown-statistics', component_property='value')
)
def update_input_container(selected_stat):
    if selected_stat == 'Yearly Statistics':
        return False
    else:
        return True 

# 2. Callback para actualizar el contenedor de salida y generar los gráficos
@app.callback(
    Output(component_id='output-container', component_property='children'),
    [Input(component_id='dropdown-statistics', component_property='value'),
     Input(component_id='select-year', component_property='value')]
)
def update_output_container(report_type, selected_year):
    if report_type == 'Recession Period Statistics':
        recession_data = df[df['Recession'] == 1]
        return [
            dcc.Graph(figure=create_graph(recession_data, 'plot_1')),
            dcc.Graph(figure=create_graph(recession_data, 'plot_2')),
            dcc.Graph(figure=create_graph(recession_data, 'plot_3')),
            dcc.Graph(figure=create_graph(recession_data, 'plot_4'))
        ]
    elif report_type == 'Yearly Statistics':
        year_data = df[df['Year'] == int(selected_year)]
        return [
            dcc.Graph(figure=create_graph(year_data, 'plot_1')),
            dcc.Graph(figure=create_graph(year_data, 'plot_2')),
            dcc.Graph(figure=create_graph(year_data, 'plot_3')),
            dcc.Graph(figure=create_graph(year_data, 'plot_4'))
        ]

# Correr la aplicación
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
# Task 2.5: Create and display graphs for Recession Report Statistics

import requests
import io
import pandas as pd
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px

# URL del archivo CSV
URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/historical_automobile_sales.csv"

# Obtener los datos de la URL
response = requests.get(URL)
response.raise_for_status()  # Levantar un error si la solicitud falló

# Convertir el contenido de la respuesta en un formato legible para pandas
csv_content = io.StringIO(response.text)

# Leer los datos CSV en un dataframe de pandas
df = pd.read_csv(csv_content)

# Filtrar los datos de recesión
recession_data = df[df['Recession'] == 1]

# Función para crear gráficos utilizando los datos
def create_graph(data, plot_id):
    fig = px.bar(data, x='Year', y='Automobile_Sales', title=f'Sales for {plot_id}')
    return fig

# Crear la aplicación Dash
app = dash.Dash(__name__)

# Lista de años para el menú desplegable de años
year_list = [i for i in range(1980, 2024)]

# Definir el layout de la aplicación
app.layout = html.Div(
    children=[
        # Título de la aplicación con estilo
        html.H1(
            "Automobile Sales Statistics Dashboard",
            style={
                'textAlign': 'center',
                'color': '#503D36',
                'font-size': '24px'
            }
        ),
        
        # Menú desplegable para seleccionar el tipo de informe
        dcc.Dropdown(
            id='dropdown-statistics',
            options=[
                {'label': 'Yearly Statistics', 'value': 'Yearly Statistics'},
                {'label': 'Recession Period Statistics', 'value': 'Recession Period Statistics'}
            ],
            placeholder='Select a report type',
            value='Select Statistics',
            style={
                'width': '80%',
                'padding': '3px',
                'font-size': '20px',
                'text-align-last': 'center'
            }
        ),
        
        # Menú desplegable para seleccionar el año
        dcc.Dropdown(
            id='select-year',
            options=[{'label': str(i), 'value': str(i)} for i in year_list],
            placeholder='Select year',
            value='Select-year',
            style={
                'width': '80%',
                'padding': '3px',
                'font-size': '20px',
                'text-align-last': 'center'
            }
        ),
        
        # División para mostrar la salida (output)
        html.Div(
            [
                html.Div(
                    id='output-container',
                    className='chart-grid',
                    style={'display': 'flex'}
                )
            ]
        )
    ]
)

# Callback para actualizar el contenedor de entrada
@app.callback(
    Output(component_id='select-year', component_property='disabled'),
    Input(component_id='dropdown-statistics', component_property='value')
)
def update_input_container(selected_stat):
    if selected_stat == 'Yearly Statistics':
        return False
    else:
        return True 

# Callback para actualizar el contenedor de salida y generar los gráficos
@app.callback(
    Output(component_id='output-container', component_property='children'),
    [Input(component_id='dropdown-statistics', component_property='value'),
     Input(component_id='select-year', component_property='value')]
)
def update_output_container(report_type, selected_year):
    if report_type == 'Recession Period Statistics':
        # Plot 1: Automobile sales fluctuate over Recession Period (year wise) using line chart
        yearly_rec = recession_data.groupby('Year')['Automobile_Sales'].mean().reset_index()
        R_chart1 = dcc.Graph(
            figure=px.line(yearly_rec, 
                x='Year', 
                y='Automobile_Sales', 
                title="Automobile Sales Fluctuation During Recession Period")
        )

        # Plot 2: Average number of vehicles sold by vehicle type (Bar chart)
        average_sales = recession_data.groupby('Vehicle_Type')['Automobile_Sales'].mean().reset_index()
        R_chart2 = dcc.Graph(
            figure=px.bar(average_sales, 
                x='Vehicle_Type', 
                y='Automobile_Sales', 
                title="Average Automobile Sales by Vehicle Type During Recession")
        )

        # Plot 3: Total expenditure share by vehicle type during recessions (Pie chart)
        exp_rec = recession_data.groupby('Vehicle_Type')['Advertising_Expenditure'].sum().reset_index()
        R_chart3 = dcc.Graph(
            figure=px.pie(exp_rec,
                values='Advertising_Expenditure', 
                names='Vehicle_Type',
                title="Total Advertising Expenditure Share by Vehicle Type During Recession")
        )

        # Plot 4: Effect of unemployment rate on vehicle type and sales (Bar chart)
        unemp_data = recession_data.groupby(['unemployment_rate', 'Vehicle_Type'])['Automobile_Sales'].mean().reset_index()
        R_chart4 = dcc.Graph(
            figure=px.bar(unemp_data, 
                x='unemployment_rate', 
                y='Automobile_Sales', 
                color='Vehicle_Type',
                labels={'unemployment_rate': 'Unemployment Rate', 'Automobile_Sales': 'Average Automobile Sales'},
                title="Effect of Unemployment Rate on Vehicle Type and Sales During Recession")
        )

        return [
            html.Div(className='chart-item', children=[html.Div(children=R_chart1), html.Div(children=R_chart2)], style={'display': 'flex'}),
            html.Div(className='chart-item', children=[html.Div(children=R_chart3), html.Div(children=R_chart4)], style={'display': 'flex'})
        ]
    elif report_type == 'Yearly Statistics':
        year_data = df[df['Year'] == int(selected_year)]
        return [
            dcc.Graph(figure=create_graph(year_data, 'plot_1')),
            dcc.Graph(figure=create_graph(year_data, 'plot_2')),
            dcc.Graph(figure=create_graph(year_data, 'plot_3')),
            dcc.Graph(figure=create_graph(year_data, 'plot_4'))
        ]

# Correr la aplicación
if __name__ == '__main__':
    app.run_server(debug=True)
```
